```python
import requests as req
from time import sleep
import os, re
from bs4 import BeautifulSoup as bs
```

```python
res = req.get('https://www.gutenberg.org/browse/languages/zh')
soup = bs(res.text, "lxml")
elms = soup.select('div.pgdbbylanguage ul li a')
for index, a in enumerate(elms):
    if index == 6:
        break
    aTitle = a.get_text()
    aLink = 'https://www.gutenberg.org' + a ['href']
    m = re.search(r'\/ebooks\/(\d+)', a['href'])
    sn = m[1]
    print(aTitle)
    print(aLink)
    print(sn)
    print(f'https://www.gutenberg.org/cache/epub/{sn}/pg{sn}.txt')
    link_txt = f'https://www.gutenberg.org/cache/epub/{sn}/pg{sn}.txt'
    res_ = req.get(link_txt)
    list_cht = re.findall(r'[\u4E00-\u9FFF]+',res_.text)
    print(' . '.join(list_cht))
    print('=' * 50)
    sleep(0.5)
```
